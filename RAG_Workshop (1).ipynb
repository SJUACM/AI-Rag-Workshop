```python
from sentence_transformers import SentenceTransformer
import openai
import requests
import json
import numpy as np
from google.colab import userdata
```

```python
openrouter_api_key = userdata.get('OPENROUTER_API_KEY')
```

```python
# Load a Hugging Face model
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')  # Specify the model name you want to use
```

```python
def get_embedding(sentence):
    return model.encode(sentence)

def get_cosine_similarity(embedding1, embedding2):
    return np.dot(embedding1, embedding2) / (np.linalg.norm(embedding1) * np.linalg.norm(embedding2))

# Function to split the text into paragraphs
def split_into_paragraphs(text):
    paragraphs = text.split('\n\n')
    # Filter out any empty paragraphs or ones that are too short to be meaningful
    return [para for para in paragraphs if len(para) > 40]


def get_most_relevant_embeddings(embeddings, query_embedding):
    similarities = []
    for e in embeddings:
        sim = get_cosine_similarity(query_embedding, embeddings[e])
        similarities.append((sim, e))

    # Sort by similarity in descending order
    similarities.sort(reverse=True, key=lambda x: x[0])

    # Get top 10 similarities
    return similarities[:10]
```

```python
sentence = "Leslie University completed the Headstarter course"

sentence2 = "Collins worked at Samsung"

embedding1 = get_embedding(sentence)
embedding2 = get_embedding(sentence2)
```

```python
cosine_similarity = get_cosine_similarity(embedding1, embedding2)

print(cosine_similarity)
```

```python
with open("textForRAG.txt", 'r') as file:
    text_contents = file.read()
```

```python
print(text_contents)
```

```python
paragraphs = split_into_paragraphs(text_contents)
```

```python
len(paragraphs)
```

```python
for p in paragraphs:
    print(p)
    print('-' * 50)
    print()
```

```python
embeddings = get_embedding(paragraphs)
```

```python
embeddings[0]
```

```python
embeddings_context = {}

for k in paragraphs:
    embeddings_context[k] = get_embedding(k)
```

```python
q = input("What is your question?")

q_embedding = get_embedding(q)

relevant_embeddings = get_most_relevant_embeddings(embeddings_context, q_embedding)

for e in relevant_embeddings:
    print(e)
```

```python
prompt = """You are a helpful assistant. Answer any questions I have in a concise way. You always answer questions with only the context that you have been provided.

CONTEXT:
"""
```

```python
for r in relevant_embeddings[:5]:
    prompt += f"\n{r[1]}"
```

```python
print(prompt)
```

```python

response = requests.post(
  url="https://openrouter.ai/api/v1/chat/completions",
  headers={
    "Authorization": f"Bearer {openrouter_api_key}",
  },
  data=json.dumps({
    "model": "google/gemma-7b-it:free",
    "messages": [
      {"role": "user", "content": prompt}
    ]
  })
)


llm_response = response.json().get("choices")[0].get('message').get("content")
print(llm_response)
```
